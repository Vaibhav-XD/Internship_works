```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen as uReq
import re
```

```python
url = " https://yoshops.com"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup
```

```python
categories = soup.find_all('li',class_='dropdown')[1:]
```

```python
categories
```

```python
from collections import Counter
```

```python
data = []

for category in categories:
  category_name = category.find('a').text.strip()
  sub_category = category.find_all('ul',class_='dropdown-menu')
  print(f"category: {category_name}")
  
  for subcategory in sub_category:
    sub_cat_name = subcategory.find_all('li')
    
    for name in sub_cat_name:
      print(f"\tsubcategories:{name.text.strip()}")
      data.append({'category': category_name,'subcategory':name.text.strip()})
df = pd.DataFrame(data)
```
