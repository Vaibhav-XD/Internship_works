```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.backends.backend_pdf import PdfPages
```

```python
order['Order Date and Time Stamp'] = pd.to_datetime(order['Order Date and Time Stamp'])
order['Year']= order['Order Date and Time Stamp'].dt.year
order['month']= order['Order Date and Time Stamp'].dt.month
```

```python
pdf = PdfPages('myplots.pdf')
#Enter 1 to see the analysis of Reviews given by Customers
def analyze_review():
  fig = plt.figure()
  star_index = review['stars'].value_counts().keys()[:5]
  star_value = review['stars'].value_counts()[:5]
  fig, ax = plt.subplots()
# Plot the distribution of review scores with added space between bars
  ax.bar(star_index, star_value, width=0.5)
  ax.set_title('Distribution of Review Scores')
  ax.set_xlabel('Stars')
  ax.set_ylabel('freq')
  pdf.savefig(fig)  # saves the current figure into a pdf page
  plt.show()
  
#Enter 2 to see the analysis of different payment methods used by the Customers
def analyze_payment():
    fig = plt.figure()
    payment_index = order['Payment Method'].value_counts().keys()[:3]
    payment_value = order['Payment Method'].value_counts()[:3]
    fig, ax = plt.subplots()
    # Plot the distribution of review scores with added space between bars
    ax.bar(payment_index, payment_value, width=0.5)
    ax.set_title('analysis of different payment methods used by the Customers')
    ax.set_xlabel('Payment_Method')
    ax.set_ylabel('freq')
    pdf.savefig(fig)  # saves the current figure into a pdf page
    
    plt.show()
    
#5. Enter 3 to see the analysis of Top Consumer States of India
def analyze_State():
  State_index = order['Shipping State'].value_counts().keys()[:5]
  State_value = order['Shipping State'].value_counts()[:5]
  fig, ax = plt.subplots()
# Plot the distribution of review scores with added space between bars
  ax.bar(State_index, State_value, width=0.5)
  ax.set_title('analysis of Top Consumer States of India')
  ax.set_xlabel('Consumer state')
  ax.set_ylabel('freq')  
  pdf.savefig(fig)  # saves the current figure into a pdf page
  plt.show()
  
#6. Enter 4 to see the analysis of Top Consumer Cities of India
def analyze_City():
  City_index = order['Shipping City'].value_counts().keys()[:5]
  City_value = order['Shipping City'].value_counts()[:5]
  fig, ax = plt.subplots()
# Plot the distribution of review scores with added space between bars
  ax.bar(City_index, City_value, width=0.5)
  ax.set_title('analysis of Top Consumer States of India')
  ax.set_xlabel('Consumer state')
  ax.set_ylabel('freq')  
  pdf.savefig(fig)  # saves the current figure into a pdf page
  plt.show()
  
#7. Enter 5 to see the analysis of Top Selling Product Categories
def analyze_Product():
  LineItem_index = order['LineItem Name'].value_counts().keys()[:2]
  
  LineItem_value = order['LineItem Name'].value_counts()[:2]
  
  fig, ax = plt.subplots()
# Plot the distribution of review scores with added space between bars
  ax.bar(LineItem_index, LineItem_value, width=0.5)
  ax.set_title('analysis of Top Selling Product Categories')
  ax.set_xlabel('Product Categories')
  ax.set_ylabel('freq')
  pdf.savefig(fig)  # saves the current figure into a pdf page
  
  plt.show()
  
  
#8. Enter 6 to see the analysis of Reviews for All Product Categories
def analyze_category():
  reviewed_category = review.groupby('category')['stars'].count()
  fig, ax = plt.subplots()
  
  reviewed_category.plot.bar(figsize=(16,5))
  plt.title(' analysis of Reviews for All Product Categories')
  plt.xlabel('ALL product Category')
  plt.ylabel('Reviews')   
  pdf.savefig(fig)  # saves the current figure into a pdf page
    
  plt.show()
    
#9. Enter 7 to see the analysis of Number of Orders Per Month Per Year
def analyze_No_order_YM():
    fig = plt.figure()
    orders_by_month_year = order.groupby(['Year', 'month'])['Order #'].count()
    orders_by_month_year.plot.bar(figsize=(15,5))
    plt.title('Number of Orders Per Month Per Year')
    plt.xlabel('Year, Month')
    plt.ylabel('Number of Orders')
    pdf.savefig(fig)  # saves the current figure into a pdf page
    plt.show()
    
#10. Enter 8 to see the analysis of Reviews for Number of Orders Per Month Per Year
def analyze_r_No_order_YM():
  fig = plt.figure()
  reviews_by_year_month = review.groupby([order['Order Date and Time Stamp'].dt.year,
                                          order['Order Date and Time Stamp'].dt.month])['stars'].count()
  reviews_by_year_month.plot.bar(figsize=(10,6))
  plt.title('Reviews of Orders Per Month Per Year')
  plt.xlabel('Number of Orders Per Month Per Year')
  plt.ylabel('Reviews')
  pdf.savefig(fig)  # saves the current figure into a pdf page 
  plt.show()
  
#11. Enter 9 to see the analysis of Number of Orders Across Parts of a Day
def analyze_part_of_day():
    fig = plt.figure()
    bins=[0, 6, 12, 18, 24]
    order['hour']=order['Order Date and Time Stamp'].dt.hour
    labels = ['Night', 'Morning', 'Afternoon', 'Evening']
    order['part_of_the_day'] = pd.cut(order['hour'], bins=bins, labels=labels, include_lowest=True)
    order_part_of_the_day = order.groupby('part_of_the_day')['Order #'].count()
    order_part_of_the_day.plot.bar(figsize=(10,6))
    plt.title('analysis of Number of Orders Across Parts of a Day')
    plt.xlabel('Number of orders across part of a day')
    plt.ylabel('freq')
    pdf.savefig(fig)  # saves the current figure into a pdf page
    
    plt.show()
    pdf.close()
    
def all_fuction():
  analyze_review()
  analyze_payment()
  analyze_State()
  analyze_City()
  analyze_Product()
  analyze_category()
  analyze_No_order_YM()
  analyze_r_No_order_YM()
  analyze_part_of_day()
    
choice = input("Enter a number in range of 1-10 : ") # put input in the code if you want to execute by in range 
if choice == '1':
    analyze_review()
elif choice == '2':
    analyze_payment()
elif choice == '3':
    analyze_State()
elif choice == '4':
    analyze_City()
elif choice == '5':
    analyze_Product()
elif choice == '6':
    analyze_category()
elif choice == '7':
    analyze_No_order_YM()
elif choice == '8':
    analyze_r_No_order_YM()
elif choice == '9':
    analyze_part_of_day()
elif choice == '10':
    all_fuction()
else:
    print("Invalid choice. Please enter a number between 1 and 10.")
```
