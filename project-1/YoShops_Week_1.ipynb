1. Python program to Create a excel file 


```python
#Create an Excel Sheet
writer = pd.ExcelWriter('demo.xlsx', engine='xlsxwriter')
writer.close()
```

```python
#Add Bulk Data to an Excel Sheet
# dataframe Name and Age columns
df = pd.DataFrame({'Name': ['A', 'B', 'C', 'D'],'Age': ['great', 0, 30, 50]})

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('demo.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df.to_excel(writer, sheet_name='Sheet1', index=False)

# Close the Pandas Excel writer and output the Excel file.
writer.close()
```

```python
#Read Data From an Excel Sheet
reader = openpyxl.load_workbook('demo.xlsx')
ws = reader['Sheet1']
```

```python

# Create a bar chart for weekly, monthly, and yearly reports
fig, ax = plt.subplots(3, 1, figsize=(8, 12))

# Plot weekly report
ax[0].bar(survey["Week"].index, survey["Week"].values)
ax[0].set_title('Weekly Report')

# Plot monthly report
ax[1].bar(survey["month"].index, survey["month"].values)
ax[1].set_title('Monthly Report')

# Plot yearly report
ax[2].bar(survey["year"].index, survey["year"].values)
ax[2].set_title('Yearly Report')

# Set x-axis tick labels
for axis in ax:
    axis.set_xticklabels(axis.get_xticklabels(), rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
